```python
import os
import sys
import urllib.request
import urllib.parse
import json
import re
import matplotlib.pyplot as plt
from PIL import Image

# 이미지 열기
ImageList = [
    Image.open('/content/drive/MyDrive/회사/DNB_kt.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_media.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_ai.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_bigdata.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_cloud.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_bigtech.jpg')
]

# 네이버 오픈API 계정
client_id = "zOEa21z01f_GWjH6b9dj"
client_secret = "6cHPjStXeu"

# 검색 텍스트
TextList = ['KT', 'IPTV', 'AI 인공지능', '빅데이터', '클라우드', '빅테크']

# 표지 출력
title = Image.open('/content/drive/MyDrive/회사/DNB_title.jpg').resize((1120,280))
plt.figure(figsize = (11,2))
plt.imshow(title)
plt.axis('off')
plt.show()

for Img, Text in zip(ImageList, TextList):
  # 원본 크기 이미지 가져오기
  width, height = Img.size

  # 이미지 크기 조정
  height = int((800/width) * height)
  Img = Img.resize((800, height))

  fig, ax = plt.subplots(figsize = (11, height/100))
  plt.imshow(Img)
  plt.axis('off')
  plt.show()

  encText = urllib.parse.quote(Text)
  url = "https://openapi.naver.com/v1/search/news?query=" + encText + "&display=10&sort=date"

  request = urllib.request.Request(url)
  request.add_header("X-Naver-Client-Id",client_id)
  request.add_header("X-Naver-Client-Secret",client_secret)

  response = urllib.request.urlopen(request)
  rescode = response.getcode()

  if rescode == 200:
      response_body = response.read()
      result = json.loads(response_body.decode('utf-8'))

      items = result.get('items',[])
      for i, item in enumerate(items, 1):
        title_raw = item['title']
        title = re.sub(r'<.*?>','',title_raw)
        link = item['link']
        print(f"   ◾\033[1m{title}\033[0m")
        print(f"   {link}")

  else:
      print("Error Code:", rescode)
```

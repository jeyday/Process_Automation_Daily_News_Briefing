```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import matplotlib.pyplot as plt
from PIL import Image

# 당일 날짜 출력
now = datetime.now()

# 이미지 열기
image_list =[
    Image.open('/content/drive/MyDrive/회사/DNB_kt.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_media.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_aibigdata.jpg')
]

# 검색하고자 하는 키워드 설정
keyword_list = ['kt', '미디어 산업', 'AI 빅데이터']

# 뉴스기사 크롤링
def daily_briefing(keyword):
  url = f'https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={keyword}'
  headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.864.59 Safari/537.36 Edg/91.0.864.59'}
  search = requests.get(url,headers = headers)

  soup = BeautifulSoup(search.text, 'html.parser')

  # 뉴스기사 리스트 가져오기
  news_list = soup.select('#main_pack > section > div > div.group_news > ul > li')

  count = 0
  for article in news_list:
    if count >= 7:
      break
    articles = article.select_one('a.news_tit')
    print(f"   ◾\033[1m{articles.text}\033[0m")
    print("  ", articles['href'])
    count += 1


# 출력
plt.imshow(Image.open('/content/drive/MyDrive/회사/DNB_title.jpg'))
plt.axis('off')

for img, keyword in zip(image_list, keyword_list):
  fig, ax = plt.subplots()
  plt.imshow(img)
  plt.axis('off')  # 축 숨기기
  plt.show()
  daily_briefing(keyword)
  print()
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import matplotlib.pyplot as plt
from PIL import Image

# 이미지 열기
image_list = [
    Image.open('/content/drive/MyDrive/회사/DNB_kt.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_media.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_ai.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_bigdata.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_cloud.jpg'),
    Image.open('/content/drive/MyDrive/회사/DNB_bigtech.jpg')
]

# 검색하고자 하는 키워드 설정
keyword_list = ['kt', '미디어 산업', 'AI', '빅데이터', '클라우드', '빅테크 양자컴퓨터']

# 기사 제목 40자 내외로 줄이기
def shorten_title(title):
    if len(title) > 59:
        return f'{title[:56]}...'
    return title

# 뉴스기사 크롤링
def daily_briefing(keyword):
    url = f'https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query={keyword}'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.864.59 Safari/537.36 Edg/91.0.864.59'}
    search = requests.get(url, headers=headers)

    soup = BeautifulSoup(search.text, 'html.parser')

    # 뉴스기사 리스트 가져오기
    news_list = soup.select('#main_pack > section > div > div.group_news > ul > li')

    count = 0
    for article in news_list:
        if count >= 5:
            break
        articles = article.select_one('a.news_tit')
        print(f'   ◾\033[1m{(articles.text)}\033[0m')
        print('  ', articles['href'])
        count += 1

# 출력
resized_title = Image.open('/content/drive/MyDrive/회사/DNB_title.jpg').resize((1120, 280))
plt.figure(figsize=(11, 2))
plt.imshow(resized_title)
plt.axis('off')

for img, keyword in zip(image_list, keyword_list):
    # 개별 이미지 원본 크기 가져오기
    orig_width, orig_height = img.size

    # 가로 크기 1000px로 조정, 세로 크기는 비율 유지
    new_height = int((800 / orig_width) * orig_height)
    resized_img = img.resize((800, new_height))

    # 이미지 크기에 맞게 figsize 조정
    fig, ax = plt.subplots(figsize=(11, new_height / 100))  # figsize에서 세로 길이 조정
    plt.imshow(resized_img)
    plt.axis('off')  # 축 숨기기
    plt.show()
    daily_briefing(keyword)
    print()

# 해외 뉴스 기사(임시)
img_overseas = Image.open('/content/drive/MyDrive/회사/DNB_global.jpg')
orig_width, orig_height = img_overseas.size

# 가로 크기를 1000px로 조정, 세로 크기는 비율 유지
new_height = int((800 / orig_width) * orig_height)
img_overseas = img_overseas.resize((800, new_height))

plt.figure(figsize=(11, new_height / 100))  # 해외 뉴스 기사도 동일하게 figsize 조정
plt.imshow(img_overseas)
plt.axis('off')
plt.show()

print(f'   ◾\033[1m{shorten_title("Turkcell broke new ground in the world in quantum key distribution")}\033[0m')
print(f'   https://www.donanimhaber.com/turkcell-kuantum-anahtar-dagitiminda-dunyada-bir-ilke-imza-atti--188213')  # 튀르키예

print(f'   ◾\033[1m{shorten_title("Digital transformation cooperation between Quang Binh Provincial People Committee and Viettel Group")}\033[0m')
print(f'   https://www.donanimhaber.com/turkcell-kuantum-anahtar-dagitiminda-dunyada-bir-ilke-imza-atti--188213')
```
